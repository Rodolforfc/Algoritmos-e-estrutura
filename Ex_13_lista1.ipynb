```python
# Aluno: Rodolfo
# Tendo como dado de entrada a altura (h) de uma pessoa, 
#construa um algoritmo que calcule seu peso ideal, utilizando as seguintes fórmulas:
# Para homens: (72.7*h) - 58
# Para mulheres: (62.1*h) - 44.7

sexo = str(input('Digite seu gênero, [M] masculino [F] feminino')) 
H = float(input('Digite sua altura'))
peso_m = (72.7 * H) - 58
peso_f = (62.1 * H) - 44.7
if sexo == "M": 
 print('O valor do peso ideal para altura masculina informada é:', peso_m)
if sexo == "F":
 print('O valor do peso ideal para altura feminina informada é:', peso_f)
else: 
 print('Digite um valor válido')
```
